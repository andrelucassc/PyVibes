```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1.7 Elemento mola

Mola é um tipo de elo mecânico. Uma força é desenvolvida na mola sempre que houver movimento relativo entre suas extremidades. Dado pela fórmula:  

$ F = k \times x $

```python
def plotar_deslocamento(constante):
    df = pd.DataFrame(np.linspace(0, 15,30), columns=['deslocamento'])
    df['força'] = df['deslocamento']*constante
    return sns.lineplot(x='deslocamento', y='força', data=df)
```

```python
plotar_deslocamento(constante=2)
```
